```python
data=df.to_dict(orient="records")
data
```

```python
# we can't add nan values in to database so we need to replace nan values with None
cleaned_data = []
for record in data:
    # Replace NaN with None
    cleaned_record = {key: (None if np.isnan(value) else value) if isinstance(value, float) else value for key, value in record.items()}
    cleaned_data.append(cleaned_record)

cleaned_data
```

```python
# Convert cleaned_data (list) to a DataFrame

df = pd.DataFrame(cleaned_data)

# Convert to a list of dictionaries
data = df.to_dict(orient="records")
```
